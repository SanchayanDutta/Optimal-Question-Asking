```python
import matplotlib.pyplot as plt
import numpy as np

# Data: each list represents the remaining number of options at each step.
gpt_o3     = [100, 45, 25, 14, 6, 3, 2]
gemini     = [100, 45, 23, 13, 7, 3, 2]
claude     = [100, 52, 24, 21, 15, 8, 4, 3, 2]
grok       = [100, 45, 25, 22, 15, 10, 7, 2]
oracle     = [100, 51, 21, 8, 5, 2]

# Function to compute entropy in bits (log2 of the number of options).
def entropy_in_bits(options_list):
    return [np.log2(x) for x in options_list]

# Prepare the data for plotting: x-values are steps, y-values are log2(N).
models = {
    "GPT o3": gpt_o3,
    "Gemini 2.5 Pro": gemini,
    "Claude 3.7": claude,
    "Grok": grok,
    "Oracle": oracle
}

plt.figure(figsize=(8, 5))

for label, data in models.items():
    x_values = range(1, len(data) + 1)
    y_values = entropy_in_bits(data)
    plt.plot(x_values, y_values, marker='o', label=label)

plt.title("100 Animals Dataset: Entropy (in bits) Across Steps")
plt.xlabel("Step")
plt.ylabel("Entropy = log2(# of Remaining Options)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
